# Task 1: Predict the percentage of a student based on the no. of study hours.

importing the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

reading the data set

```python
link="http://bit.ly/w-data"
data=pd.read_csv(link)
print("data sucessfully imported")
```

```python
data.head()
```

Plot the graphs to find the relationship between variables present in data

```python
sns.regplot(x="Hours",y="Scores",data=data)
```

```python
sns.jointplot(x="Hours",y="Scores",data=data)
```

```python
plt.hist(x="Hours",data=data)
plt.show()
```

describe the data

```python
data.describe()
```

Dividing the dataset into labels(output) and attributes(input)

```python
x=data.iloc[:,0:1].values
y=data.iloc[:,1].values
```

Next step is to split this data to training and testing sets.We use scikit-learn built-in method train_test_split()

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Training the algorithm

We have to train the algorithm and check the results on testing set

```python
from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(x_train,y_train)
print("Training complete")
```

Plot the regression line in the scatter plot of original data

line=r.coef_*x+r.intercept_
plt.scatter(x,y)
plt.plot(x,line)
plt.show()

# Making Predictions

We have trained our algorithm, now it's the time for testing

```python
y_pred=r.predict(x_test)
y_pred
```

Compare actual values with predicted values

df=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
df

Customizing the input

```python
hours=[[9.25]]
s_pred=r.predict(hours)
s_pred
```

# Evaluate the model

The last step is to evaluate the performance of algorithm.We can evaluate by calculating the mean squared error or mean absolute error.There are many such metrics.

Mean Absolute Error

```python
from sklearn import metrics
print("mean absoloute error:",metrics.mean_absolute_error(y_test,y_pred))
```

Mean Squared Error

```python
print("mean squared error:",metrics.mean_squared_error(y_test,y_pred))
```
